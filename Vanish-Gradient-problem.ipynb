```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Social_Network.csv')
```

```python
df.head()
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
## Input Layer
x=df.iloc[:,:2]
y=df.iloc[:,-1]
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x=scaler.fit_transform(x)
```

```python
model=Sequential()
```

```python
model.add(Dense(10,activation='relu',input_dim=2))
model.add(Dense(10,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(1,activation='sigmoid')) ## yaha sigmoid esliye liya ha kyuki , vo ek classification dataset ha , agar regression hota toh linear leta
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
## Abhi dekh part yaha se start hota ha 
old_weight=model.get_weights()[0]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.20)
```

```python
value=model.fit(x_train,y_train,epochs=100)
```

```python
plt.plot(value.history['loss'])
plt.plot(value.history['accuracy'])
plt.show()
```

```python
## Ye dekh yahi pata chal raha ha ki vanishing gradient descent ho raha ha 
## But ye dekh meine , activation function ko sigmoid se change kareke RELU kar diya ha , toh vanish gradient problem remove ho gaya ha 
```

```python
new_weights=model.get_weights()[0]
```

```python
learning_rate=model.optimizer.get_config()['learning_rate']
```

```python
## Abhi humko gradient ka value chahiye toh. hum formula ko use karenge. wold-wnew/learning rate
```

```python
gradient=(old_weight-new_weights)/learning_rate
```
